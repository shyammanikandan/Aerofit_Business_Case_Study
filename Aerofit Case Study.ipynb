```python
def prdamt(x):
    if x['Product']=='KP281':
        x['Price']=1500
    elif x['Product']=='KP481':
        x['Price']=1750
    else:
        x['Price']=2500
    return x
```

```python
data=data.apply(prdamt,axis=1)
data
```

```python
def ic(x):
    if x['Income']<= 45000:
        x['Income_Category']='Low'
    elif x['Income']>45000 and x['Income']<=55000:
        x['Income_Category']='Medium'
    else:
        x['Income_Category']='High'
    return x
```

```python
data=data.apply(ic,axis=1)
```

```python
def fitnesscat(x):
    if x['Fitness']<=2:
        x['Fit_cat']='Low'
    elif x['Fitness']>2 and x['Fitness']<4:
        x['Fit_cat']='Medium'
    else:
        x['Fit_cat']='High'
    return x
        
```

```python
data=data.apply(fitnesscat,axis=1)
data
```
